```python
# o seleccionar las candidatas
df = df[['PHYSHLTH','MENTHLTH','BPHIGH4','TOLDHI2','CVDINFR4','CVDCRHD4',
         'CVDSTRK3','SEX','WEIGHT2','HEIGHT3','SMOKE100','USENOW3','ALCDAY5',
         'FRUITJU1','FRUIT1','FVBEANS','FVGREEN','FVORANG','VEGETAB1',
         'EXERANY2','PREDIAB1']]
df
```

```python
#Preparacion atributo PHYSHLTH

# Eliminar filas que generan ruido
df=df[df['PHYSHLTH'] != 77.0]
df=df[df['PHYSHLTH'] != 99.0]

#Reemplazar nulos con el valor promedio
df['PHYSHLTH'] = df['PHYSHLTH'].fillna((df['PHYSHLTH'].mean()))

# Reemplazar los valores predefinidos del dataset de 88(ninguna) a cero
ValorNinguno = lambda x: 0 if (x==88.0) else x

#Almacenar los cambios en el nuevo atributo SaludFisica
df['SaludFisica'] = df['PHYSHLTH'].apply(ValorNinguno)

# Normalización: (valor-min)/(max-min)
min_veces=0
max_veces=30
NumVeces = lambda x: (x - min_veces) / (max_veces - min_veces)

#Almacenar los cambios en el nuevo atributo SaludFisica
df['SaludFisica'] = df['SaludFisica'].apply(NumVeces)
df
```
